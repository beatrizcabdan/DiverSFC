```python
import pandas as pd
import os
```

Sensor reading (time signal) analysis to understand real data:

```python
parentFolder = 'input_data/Lanechangecitos/'

for date in os.listdir(parentFolder):
    if not os.path.isdir(parentFolder + date):
        continue

    data = pd.read_csv(parentFolder + date + "/" + date + "_merged_data_sensor.csv")
    df = pd.DataFrame(data)
    latex_code = df.describe().to_latex(
        float_format="%.1f",
        caption="Sample Table",
        label="tab:sample"
        )
    
    print(date)
    print(latex_code)
    print()
```
